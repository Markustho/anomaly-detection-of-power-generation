```python
import pandas as pd
import matplotlib.pyplot as plt

# Creating the data
data = {
    '1x SD [%]': [81.38, 76.92, 83.63, 77.96],
    '1.5 x SD [%]': [89.18, 89.07, 88.72, 89.88],
    '2 x SD [%]': [92.94, 94.45, 92.89, 95.03],
    '2.5 x SD [%]': [96.01, 96.99, 95.72, 97.80],
    '3 x SD [%]': [97.46, 98.61, 96.70, 98.79]
}

# Create a DataFrame
index = ['ML', 'ML+therm', 'ML (no RPM)', 'ML+therm (no RPM)']
df = pd.DataFrame(data, index=index)

# Display the DataFrame as a table
print(df)

# Transposing the DataFrame to have SD levels as rows and Models as columns
df_transposed = df.transpose()

# Define the custom colors
colors = ['teal', 'slategrey', 'orange', 'royalblue']

# Create subplots
plt.figure(figsize=(10, 6))

# Plotting each category (column) as a line with specified colors
for i in range(len(df.index)):
    plt.plot(df_transposed.index, df_transposed.iloc[:, i], marker='o', label=df.index[i], color=colors[i])

# Adding labels, title, and legend
plt.xlabel('SD Levels')
plt.ylabel('Values [%]')
plt.title('Comparison of Models across Different SD Levels')
plt.legend(title='Models')
plt.grid()  # Optional: Add a grid for easier readability

# Show the plot
plt.tight_layout()
plt.show()
```
